```python
from PIL import Image
import requests
from io import BytesIO
# Load the image from a URL
url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTIELwoOai4nlHgBF3KWsMIOMwMbQLGns_lRg&usqp=CAU"  # replace with your image URL
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Display the original image size
print(f"Original size: {img.size}")

# Compress and save the image
img.save("compressed_image.jpg", optimize=True, quality=10)

# Display the compressed image size
compressed_img = Image.open("compressed_image.jpg")
print(f"Compressed size: {compressed_img.size}")

# Show the compressed image
compressed_img.show()
```

```python
from PIL import Image
import requests
from io import BytesIO
import os
from IPython.display import display
# Load the image from a URL
url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTIELwoOai4nlHgBF3KWsMIOMwMbQLGns_lRg&usqp=CAU"  # replace with your image URL
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Save the original image to disk to check its size
original_path = "original_image.jpg"
img.save(original_path)

# Display the original image size in bytes
original_size = os.path.getsize(original_path)
print(f"Original file size: {original_size} bytes")

# Display the original image
print("Original Image:")
display(img)

# Compress and save the image
compressed_path = "compressed_image.jpg"
img.save(compressed_path, optimize=True, quality=20)

# Display the compressed image size in bytes
compressed_size = os.path.getsize(compressed_path)
print(f"Compressed file size: {compressed_size} bytes")

# Load and display the compressed image
compressed_img = Image.open(compressed_path)
print("Compressed Image:")
display(compressed_img)
```
