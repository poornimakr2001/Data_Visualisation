```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load a sample dataset
df = sns.load_dataset('tips')
```

```python
# Bar plot of total bill amounts
plt.figure(figsize=(8,6))
plt.bar(df['day'], df['total_bill'], color='skyblue')
plt.title('Total Bill Amount by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Boxplot of total bill by day
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=df, palette='Set3')
plt.title('Total Bill Distribution by Day')
plt.show()
```

```python
import plotly.express as px

# Scatter plot with Plotly
fig = px.scatter(df, x='total_bill', y='tip', color='day', title='Tip vs Total Bill by Day')
fig.show()
```
