```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Manually enter data for forest density in different regions (values are hypothetical)
data = np.array([
    [40, 55, 65, 70],
    [60, 75, 80, 85],
    [50, 65, 75, 90],
    [30, 50, 60, 70]
])

# Create a DataFrame for better visualization control
regions = ['Region 1', 'Region 2', 'Region 3', 'Region 4']
years = ['2021', '2022', '2023', '2024']

df = pd.DataFrame(data, index=regions, columns=years)
```

```python
# Set the plot size for better readability
plt.figure(figsize=(8, 6))

# Create a heatmap
sns.heatmap(df, annot=True, cmap='YlGnBu', linewidths=0.5, linecolor='black')

# Add titles and labels
plt.title('Forest Density in African Regions (Hypothetical Data)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Region')

# Show the plot
plt.show()
```
