```python

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/flights.csv' # Adjust the path to your file
data = pd.read_csv(file_path)

# Display basic information about the dataset
print(data.info())
print(data.head())

# Analyze flight delays
# Example: Calculate average arrival delay per airline
average_delay = data.groupby('carrier')['arr_delay'].mean().sort_values()
print(average_delay)

# Visualize the average delays
average_delay.plot(kind='bar', figsize=(12, 6))
plt.title('Average Arrival Delay per Airline')
plt.xlabel('Airline')
plt.ylabel('Average Delay (minutes)')
plt.show()

# Additional analysis: Departure delay distribution
plt.figure(figsize=(12, 6))
plt.hist(data['dep_delay'].dropna(), bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Departure Delays')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Frequency')
plt.show()

# Additional analysis: Arrival delay by month
average_delay_by_month = data.groupby('month')['arr_delay'].mean()
print(average_delay_by_month)

average_delay_by_month.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Average Arrival Delay by Month')
plt.xlabel('Month')
plt.ylabel('Average Delay (minutes)')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```
