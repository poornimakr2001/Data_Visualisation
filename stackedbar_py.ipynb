```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define a dataset
data = {
    'Manufacturer': ['Ford', 'Ford', 'Ford', 'Chevy', 'Chevy', 'Chevy', 'Toyota', 'Toyota'],
    'Model': ['Fiesta', 'Focus', 'Mustang', 'Malibu', 'Impala', 'Camaro', 'Corolla', 'Camry'],
    'Fuel_Type': ['Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Diesel'],
    'Count': [10, 15, 10, 20, 10, 15, 10, 15]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Group by Manufacturer and Fuel_Type and sum the counts
grouped_df = df.groupby(['Manufacturer', 'Fuel_Type']).sum().unstack()

# Plotting
grouped_df.plot(kind='bar', stacked=True, figsize=(10, 7))

plt.title('Stacked Bar Chart of Fuel Types by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.legend(title='Fuel Type')
plt.show()
```
