```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
categories = ['Start', 'Profit', 'Loss', 'Tax', 'Final']
values = [1000, 200, -100, -50, 1050]

# Calculate the positions and bars for the waterfall
positions = np.arange(len(categories))
cumulative = np.cumsum([0] + values[:-1])

# Convert 'values' to a NumPy array for comparison
values_array = np.array(values)  # Convert list to NumPy array
bottoms = np.where(values_array >= 0, cumulative, cumulative + values_array)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(positions, values_array, bottom=bottoms, color=['blue' if x >= 0 else 'red' for x in values_array])

# Adding labels and titles
ax.set_xticks(positions)
ax.set_xticklabels(categories)
ax.set_title('Waterfall Chart Example')
ax.axhline(0, color='black', linewidth=0.8)  # Baseline

# Show the plot
plt.show()
```
